```python
# 각 line들을 합쳐주는 함수 작성
def flatten(l):
    flatList = []
    for elem in l:
        if type(elem) == list:
            for e in elem:
                flatList.append(e)
        else:
            flatList.append(elem)
    return flatList
```

```python
word_list_okt_moon = flatten(temp_okt_moon)
print(word_list_okt_moon)
```

```python
# 영어 단어 빈도분석
import pandas as pd
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
# trump
f_eng_trump = open('C:\\Users\\bomul\\OneDrive\\바탕 화면\\23-1\\데사\\중간 대비\\3주차 연습\\trump.txt', 'r')
lines_eng_trump = f_eng_trump.readlines()[0]
f_eng_trump.close()
print(lines_eng_trump)
```

```python
# biden
f_eng_biden = open('C:\\Users\\bomul\\OneDrive\\바탕 화면\\23-1\\데사\\중간 대비\\3주차 연습\\biden.txt', 'r')
lines_eng_biden = f_eng_biden.readlines()[0]
f_eng_biden.close()
print(lines_eng_biden)
```

```python
tokenizer = RegexpTokenizer('[\w]+')
stop_words = stopwords.words('english')
```

```python
words_trump = lines_eng_trump.lower()
words_biden = lines_eng_biden.lower()

tokens_trump = tokenizer.tokenize(words_trump)
tokens_biden = tokenizer.tokenize(words_biden)

#불용어 제거
stopped_tokens_trump = [i for i in list((tokens_trump)) if not i in stop_words]
stopped_tokens_biden = [i for i in list((tokens_biden)) if not i in stop_words]

stopped_tokens2_trump = [i for i in stopped_tokens_trump if len(i) > 1]
stopped_tokens2_biden = [i for i in stopped_tokens_biden if len(i) > 1]
```
