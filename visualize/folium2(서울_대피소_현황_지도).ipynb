```python
import pandas as pd
```

```python
file = './data/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv'

#ms-office -> cp949
raw = pd.read_csv(file, encoding = 'cp949')
raw
```

```python
raw.head()
```

```python
raw.info()
```

```python
#i = 10

for i in range(len(raw)):
    lat = raw.loc[ i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']

    print(lat, long, name)
```

```python
#지도 생성
import folium
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 12)
#마커 추가
for i in range(len(raw)):
    lat = raw.loc[ i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    folium.Marker([lat, long], tooltip = name).add_to(m)
    
m
```

```python
#파일로 저장
m.save('./data/Sheltermap.html')
```

```python
from folium.plugins import MarkerCluster
```

```python
#근처 마커들끼리는 묶어서 표현
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 12)

marker_cluster = MarkerCluster().add_to(m)

for i in range(len(raw)):
    lat = raw.loc[ i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    #marker_cluster에 추가
    folium.Marker([lat, long], tooltip = name).add_to(marker_cluster)
    
m
```
