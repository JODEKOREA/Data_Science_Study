```python
#지도시각화를 위한 folium 라이브러리
import folium
```

```python
#지도 생성
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 10)
m
```

```python
#마커 추가하기(tooltip, popup)
folium.Marker([37.5536067, 126.9674308], 
              tooltip = "mouseover",
              popup = "click").add_to(m)
m
```

```python
#CircleMarker([위도, 경도]).add_to(지도이름)
folium.CircleMarker([37.5536067, 126.9674308],
                   radius = 20,
                   tooltip = 'mouseover').add_to(m)
m
```

```python
#minimap
from folium.plugins import MiniMap

m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 12)

#미니맵 추가
minimap = MiniMap()
minimap.add_to(m)
m
```
